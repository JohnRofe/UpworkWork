```python
#Import libraries for data cleaning

import pandas as pd
import numpy as np
```

```python
#Read csv 
df = pd.read_csv('output.csv')
```

```python
#Check duplicates in lsoNumber
df.duplicated(subset=['lsoNumber']).sum()
```

```python
#Print those duplicates 
df[df.duplicated(subset=['lsoNumber'], keep=False)].sort_values(by='lsoNumber')
```

```python
#Drop duplicates
df = df.drop_duplicates(subset=['lsoNumber'], keep='first')
```

```python
#drop area of law09 
df = df.drop(columns=['areaOfLaw09'])
```

```python
#drop columns language06-09
df = df.drop(columns=['language06', 'language07', 'language08', 'language09'])
```

```python
#Stripe the phone column and remove /n if any
df['phone'] = df['phone'].str.replace('\n', '')
df['phone'] = df['phone'].str.strip()
```

```python
#same for fax 
df['fax'] = df['fax'].str.replace('\n', '')
df['fax'] = df['fax'].str.strip()
```

```python
df.info()
```

```python
#Unique values in businessCity
df['businessCity'].unique()
```

```python
# Define the regular expression pattern
pattern = r'(.*[A-Z][a-z]*)([A-Z])'

# Add a 'PatternFound' label if the pattern is found in 'businessCity'
df['PatternFound'] = df['businessCity'].str.contains(pattern)

# Replace NaN values in 'PatternFound' with False
df['PatternFound'] = df['PatternFound'].fillna(False)

# Extract 'Separation' and 'businessCity' from 'businessCity' in df only if 'PatternFound' is True
df.loc[df['PatternFound'], ['Separation', 'businessCity']] = df.loc[df['PatternFound'], 'businessCity'].str.extract(pattern)

# Append 'Separation' to the end of 'businessAddress' in df only if 'PatternFound' is True
df.loc[df['PatternFound'], 'businessAddress'] = df.loc[df['PatternFound'], 'businessAddress'].str.cat(df.loc[df['PatternFound'], 'Separation'], sep=' ', na_rep='')

# Drop the 'Separation' and 'PatternFound' columns as they're no longer needed
df = df.drop(columns=['Separation', 'PatternFound'])
```

```python
#Unique values in businessCity
df['businessCity'].unique()
```
